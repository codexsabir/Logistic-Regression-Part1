```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
insurance_df = pd.read_csv('dummy_insurance_data.csv')
```

```python
insurance_df.head()
```

```python
plt.scatter(insurance_df.age,insurance_df.have_insurance,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train,X_test,y_train,y_test=train_test_split(insurance_df[['age']],insurance_df.have_insurance,train_size=0.9)
```

```python
X_train
```

```python
X_test
```

```python
 from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
